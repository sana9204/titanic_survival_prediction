### Survided Prediction

```python
da = pd.read_csv("D:/Data_Science/pandas_docs/titanic_final.csv")
da.head(2)
```

```python
da.isnull().sum()
```

```python
df['SibSp'].unique()
```

```python
x = da.drop(columns=['PassengerId','Name','Ticket','Cabin','Survived'])
```

```python
x.head(10)
```

```python
y=da['Survived']
y.head(10)
```

```python
len(y)
```

#### Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
se = da['Sex']
em = da['Embarked']
```

```python
ense = le.fit_transform(se)
enem = le.fit_transform(em)
```

```python
ense
```

```python
x['Sex']=ense
x['Embarked']=enem
```

```python
x.head(10)
```

### filling age with mean

```python
x.isnull().sum()
```

```python
x['Age'] = x['Age'].fillna(value=x['Age'].mean())
```

```python
x.isnull().sum()
```

```python
len(x)
```

#### Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
xscaler = scaler.fit_transform(x)
```

```python
xscaler
```

### Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(xscaler,y,test_size=0.2,random_state=42)
```

### Model Building

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model
```

```python
model.fit(xtrain,ytrain)
```

```python
pred=model.predict(xtest)
pred
```

```python
model.coef_
```

```python
residuals = ytest-pred
residuals
```

```python
model.intercept_
```

### Evaluation Metrics

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(ytest,pred)
```

```python
confusion_matrix(ytest,pred)
```

```python
classification_report(ytest,pred)
```

#### save model

```python
joblib.dump(model,'model.pkl')
```

```python
joblib.dump(scaler,'scaler.pkl')
```

```python
joblib.dump(le,'encoder.pkl')
```
